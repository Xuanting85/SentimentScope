```python
#Identify a problem to solve 
#Problem : To find the number of fatalities based on certain chosen features
#Source : "outbreaks.csv"
#First and foremost import Python Libraries 

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

#Loading of the dataset
outbreak_df = pd.read_csv("outbreaks.csv")
#Summarize the dataset
#Head function allows one to see the first 20 rows
print(outbreak_df.head(5))

#Missing data can be spotted in different columns such as location, species, status, etc.
#The missing values in each columns are already replaced with the value NaN

#Plot a boxplot for each column with numerical values
outbreak_df.plot(kind='line', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
#Next is to remove features that have little to not contribution and are unnecessary
#Feature Engineering
del outbreak_df['Year']
del outbreak_df['Month']
del outbreak_df['Ingredient']
del outbreak_df['Serotype/Genotype']
del outbreak_df['Food']
del outbreak_df['Status']
del outbreak_df['State']
outbreak_df.shape
```

```python
#Find the number of missing values
#Since the % of the missing values is more than 10% they have to be replaced
print(outbreak_df.isnull(). sum())

#Removing missing values or data as certain machine learning algorithms do not work with missing values
from sklearn.impute import SimpleImputer

#For missing numerical values such as hospitalizations and fatalities
#The values are not removed, however they are replaced with the median
#Median is used instead of mean as there are some outliers present hence median deals better with them
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
outbreak_df[['Hospitalizations']] = imputer.fit_transform(outbreak_df[['Hospitalizations']])

imputer = SimpleImputer(missing_values=np.nan, strategy='median')
outbreak_df[['Fatalities']] = imputer.fit_transform(outbreak_df[['Fatalities']])
```

```python
#Replacing categorical variables using one-hot encoding 
features_outbreak_df = pd.get_dummies(outbreak_df, columns=['Location','Species'])

#Remove the fatalities from the feature data 
del features_outbreak_df['Fatalities']

#print(features_outbreak_df.head(0))

#From the function shape, we can tell that there are 19119 rows and 364 columns after data preparation
features_outbreak_df.shape
```

```python
#from sklearn.model_selection import cross_val_score
from sklearn import model_selection
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.externals import joblib
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import linear_model
from sklearn import tree


#Creating X & Y arrays
X = features_outbreak_df.values
Y = outbreak_df['Fatalities']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2 , random_state=7, shuffle=True)
```

```python
#Testing various models
#Since the tested algorithms are regression models, neg absolute error is used
#A value of 0 indicates that there are no errors and it is a perfect prediction
models = []
models.append(('DTC', tree.DecisionTreeRegressor()))
models.append(('GBR', GradientBoostingRegressor()))
models.append(('LR', linear_model.LinearRegression()))
models.append(('RFR', RandomForestRegressor(n_estimators=10)))
# evaluate each model in turn
results = []
names = []
for name, model in models:	
    kfold = model_selection.KFold(n_splits=10, random_state=7)
    scoring = 'neg_mean_absolute_error'
    result = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
    results.append(result)	
    names.append(name)	
    msg = "%s: %f (%f)" % (name, result.mean(), result.std())
    print(msg)
```

```python
#Applying Grid Search to further improve the model
from sklearn.model_selection import GridSearchCV
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3 , random_state=7, shuffle=True)

model_f = GradientBoostingRegressor()

param_grid = {
    'n_estimators':[50, 100, 200],
    'max_depth':[3, 5],
    'min_samples_leaf':[10, 20],
    'loss':["ls", "huber"],
    'learning_rate' :[0.1, 0.05, 0.02]
}

gs_cv = GridSearchCV(model_f, param_grid, n_jobs=4, verbose=100)

gs_cv.fit(X_train, Y_train)

print(gs_cv.best_params_)

#{'learning_rate': 0.05, 'loss': 'ls', 'max_depth': 5, 'max_features': 'sqrt', 'min_samples_leaf': 10, 'n_estimators': 100}
```

```python
#Modify the hyper-parameters to the parameters recommended by GridSearch
#Test the modified RandomForestClassifier after GridSearch has been completed
model_t = GradientBoostingRegressor(
    learning_rate= 0.05, 
    max_depth= 5,
    max_features= None, 
    min_samples_leaf= 10, 
    loss= 'ls', 
    n_estimators= 100
)
model_t.fit(X_train, Y_train)
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(Y_train, model_t.predict(X_train))
print("Training Set Mean Absolute Error : %.4f" %mae)

mae = mean_absolute_error(Y_test, model_t.predict(X_test))
print("Test Set Mean Absolute Error : %4.f" %mae)

#Place the model into the "model" folder using joblib
joblib.dump(model_t, 'model//trained_fatalities_model.pkl')
```
